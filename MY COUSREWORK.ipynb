```python
# How to use Module 
# import random

import random as random

# Generate a random integer between 1 and 10
random_int = random.randint(1, 10)
print('Random integer:', random_int)

# Choose a random element from a list
options = ['apple', 'banana', 'cherry']
random_choice = random.choice(options)
print('Random choice:', random_choice)
```

```python
# How to use Package
# Module 1: addition.py
def add(x, y):
    "Function to add two numbers."
    return x + y

# Module 2: subtraction.py
def subtract(x, y):
    "Function to subtract two numbers."
    return x - y

# Simulating the __init__.py functionality
def init_package():
    global add, subtract

# Initialize the "package"
init_package()

# Now we can use the "package" functions
result_add = add(20, 5)
result_subtract = subtract(20, 5)

# Print the results
print("Addition Result:", result_add)
print("Subtraction Result:", result_subtract)
```

## Application of NumPy

```python
import numpy as np
a = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80])
a
```

```python
# Reshaping
a = np.arange(5,81, 5).reshape((4, 4))
print(a)
```

```python
type (a)
```

```python
# Calculating total sum, mean, standard deviation and variance of all element 
total_sum = np.sum(a)
mean = np.mean(a)
std_dev = np.std(a)
variance = np.var(a)

print("Sum of all elements:", total_sum)
print("Mean of all elements:", mean)
print("Standard deviation of all elements:", std_dev)
print("Varianceof all elements:", variance)
```

```python
# Vectorized operation: Square each all element
squared_array = a ** 2
print("Squared array:\n", squared_array)
```

## Application of Pandas

```python
import pandas as pd

# Step 1: Load your CSV file into a DataFrame (adjust the file path if needed)
df = pd.read_csv('SHEL_stock_data.csv')
df.head()

```

```python
# Check for missing values in the DataFrame
print("\nMissing Values:\n", df.isnull().sum())

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
#Add a new column for the month extracted from the 'Date' column
df['Month'] = df['Date'].dt.month
# Group by month and calculate the average of the 'Close' column
monthly_avg = df.groupby('Month')['Close'].mean()

print("Monthly Average Closing Price for Shell in 2021/2022:\n", monthly_avg)
```
